```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from palettable.cmocean.sequential import Ice_20 as palette
import scipy
```

```python
sourcetrack = pd.read_csv('~/Desktop/tracked/mixing_proportions.txt', sep='\t', header=0, index_col=0)
sourcetrack.rename(columns={'plant-associated habitat':'plant-associated'}, inplace=True)
sns_sourcetrack = (sourcetrack*100).round(2)

sns_sourcetrack = sns_sourcetrack.T[sns_sourcetrack.sum()>13]

def plot_sourcetrack(dataframe, title=False, savename=False,):
    matplotlib.rcParams.update({'font.size': 40})
    fig, ax = plt.subplots(figsize=(30,35), dpi=50)    
    data = dataframe.drop('Unknown', axis=0)
    sns.heatmap(data,
                ax=ax,
                vmin=0,
                vmax=7,
                cbar=True,
                cmap=palette.mpl_colormap,
                cbar_kws=dict(use_gridspec=False,
                              shrink=1,
                              location='top',
                             )
               )
    if savename:
        plt.savefig(savename, bbox_inches='tight')
    plt.show()
    
plot_sourcetrack(sns_sourcetrack,savename='heatmap_perc.png')
sns_sourcetrack
```
