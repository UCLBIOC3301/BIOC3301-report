```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from palettable.matplotlib import Inferno_20 as palette
import scipy
```

```python
def sort_df(dataframe):
    cols = []
    for num in range(8,38):
        cols.append('515rcbc'+str(num))
    dataframe = dataframe.reindex(reversed(cols))
    return dataframe[cols]

def plot_heatmap(dataframe, title=False, savename=False, fmt='d', annot=True, cbar=True):
    matplotlib.rcParams.update({'font.size': 40})

    fig, ax = plt.subplots(figsize=(30,40), dpi=50)
    hmp = sns.heatmap(ax=ax,
                vmin=0,
                vmax=0.6,
                data=dataframe,
                square=True,
                annot=annot,
                fmt=fmt,
                annot_kws={"size":8},
               cbar=cbar,
                      cmap=palette.mpl_colormap,
                      cbar_kws=dict(use_gridspec=False,
                              shrink=1,
                              location='top',
                             )
               )
    if title:
        plt.title(title, fontsize=40)
    if savename:
        plt.savefig(savename, bbox_inches='tight')
    plt.show()

beta_unweight = pd.read_csv('~/Desktop/rbc20_included/final/cda/bdiv_even744/weighted_unifrac_dm.txt', sep='\t', index_col=0)
beta_unweight = sort_df(beta_unweight)
plot_heatmap(beta_unweight, fmt='.4f', savename='./weighted_beta.png', cbar=True, annot=False)
```
